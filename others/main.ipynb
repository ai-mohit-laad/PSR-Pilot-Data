```python
import pandas as pd
import warnings
import datetime as dt

warnings.filterwarnings("ignore")

psr = pd.read_excel("PSR_oooi_Apr2024.xlsx")
psr = psr[["id", "airbornetouchdowntime", "arr_or_depart", "chocksoffintime", "flight_date", "flightnumber", "pltlandortkoff", "pltid"]]
psr
```

```python
psr["AcarsOUT"] = [None]*len(psr)
psr["AcarsOFF"] = [None]*len(psr)
psr["AcarsON"] = [None]*len(psr)
psr["AcarsIN"] = [None]*len(psr)

for index, row in psr.iterrows():
    if row["arr_or_depart"] == "D":
        psr["AcarsOUT"][index] = row["chocksoffintime"]
        psr["AcarsOFF"][index] = row["airbornetouchdowntime"]
    elif row["arr_or_depart"] == "A":
        psr["AcarsON"][index] = row["airbornetouchdowntime"]
        psr["AcarsIN"][index] = row["chocksoffintime"]

psr["flight_date"] = psr["flight_date"].dt.strftime("%d%b%Y")

psr
```

```python
mm = pd.read_excel("ai_psr_01Apr2024_30Apr2024.xlsx")

mm["flightnumber"] = [None]*len(mm)

for index, row in mm.iterrows():
    if row["FLIGHT"][0] == '0':
        mm["flightnumber"][index] = 'AI' + row["FLIGHT"][1:].strip()
    else:
        mm["flightnumber"][index] = 'AI' + row["FLIGHT"].strip()

mm["flight_date"] = mm["FLTDATE"].dt.strftime("%d%b%Y")
mm
```

```python
psr["flightnumber"] = psr["flightnumber"].astype(str)
psr["flight_date"] = psr["flight_date"].astype(str)

mm["flightnumber"] = mm["flightnumber"].astype(str)
mm["flight_date"] = mm["flight_date"].astype(str)
```

```python
psr.to_excel("psr.xlsx")
mm.to_excel("mm.xlsx")
```

```python
# do a left join on psr df to mm df based on flightnumber and flight_date columns 
df = pd.merge(psr, mm, how='left', on=["flightnumber", "flight_date"])
df.to_excel("output.xlsx")
```
